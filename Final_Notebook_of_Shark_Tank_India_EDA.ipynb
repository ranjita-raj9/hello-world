# **BASIC  INFORMATION**

```python
df['Season Number'] = df['Season Number'].astype(pd.Int32Dtype())
df['Episode Number'] = df['Episode Number'].astype(pd.Int32Dtype())
df['Pitch Number'] = df['Pitch Number'].astype(pd.Int32Dtype())

df['Number of Presenters'] = df['Number of Presenters'].astype(pd.Int32Dtype())
df['Male Presenters'] = df['Male Presenters'].astype(pd.Int32Dtype())
df['Female Presenters'] = df['Female Presenters'].astype(pd.Int32Dtype())
df['Transgender Presenters'] = df['Transgender Presenters'].astype(pd.Int32Dtype())
df['Couple Presenters'] = df['Couple Presenters'].astype(pd.Int32Dtype())
df['Gross Margin'] = df['Gross Margin'].astype(pd.Int32Dtype())

df['Started in'] = df['Started in'].astype(pd.Int32Dtype())
df['Yearly Revenue'] = df['Yearly Revenue'].astype(pd.Int32Dtype())

df['Received Offer'] = df['Received Offer'].astype(pd.Int32Dtype())
df['Accepted Offer'] = df['Accepted Offer'].astype(pd.Int32Dtype())
```

```python
df_season1 = df.loc[df['Season Number']==1]
df_season1_without_unseen = df.loc[(df['Season Number']==1) & (df['Episode Number']!=0)]
df_season2 = df.loc[df['Season Number']==2]
df_season3 = df.loc[(df['Season Number']==3) | (df['Season Number'].isnull())]
```

```python
# Season-wise number of episodes
pd.pivot_table(df, values='Episode Number', columns='Season Number', aggfunc='max')
```

```python
print('Total seasons in Indian SharkTank:',df['Season Number'].max())
print("Startups came for pitching :",df['Pitch Number'].max())
print("In Season 1, in", df_season1['Episode Number'].max(), "episodes, there were", df_season1.loc[df_season1['Episode Number']!=0]['Startup Name'].count(), "(real) pitches and", df_season1.loc[df_season1['Episode Number']==0]['Startup Name'].count(),"unseen pitches")
print("In Season 2, in", df_season2['Episode Number'].max(), "episodes, there were", df_season2.loc[df_season2['Episode Number']!=0]['Startup Name'].count(), "(real) pitches and", df_season2.loc[df_season2['Episode Number']==0]['Startup Name'].count(),"unseen pitch")
print("In Season 3, in", df_season3['Episode Number'].max(), "episodes, there were", df_season3.loc[df_season3['Episode Number']!=0]['Startup Name'].count(), "(real) pitches")
```
